```python
import PyPDF2
from fixthaipdf import clean

# ฟังก์ชันสำหรับลบอักขระที่ไม่ต้องการ
def clean_text(text):
    # กำหนดการแทนที่อักขระที่ผิดพลาด
    replacements = {
        "É": "",
        "Ê": "",
        "ï": "",
        "®": "",
        # สามารถเพิ่มอักขระที่ผิดพลาดเพิ่มเติมได้ที่นี่
    }
    # แทนที่อักขระที่ผิดพลาด
    for key, value in replacements.items():
        text = text.replace(key, value)
    return text

# เปิดไฟล์ PDF ในโหมดอ่านไบนารี่
with open(r"D:\EasyOCR\Calendar_KPS_2567.pdf", 'rb') as pdf_file:
    # สร้างตัวอ่านไฟล์ PDF
    reader = PyPDF2.PdfReader(pdf_file)
    # วนลูปผ่านทุกหน้าในไฟล์ PDF
    for page_num in range(len(reader.pages)):
        # ดึงหน้าแต่ละหน้า
        page = reader.pages[page_num]
        # ดึงข้อความจากหน้า
        text = page.extract_text()
        # ทำความสะอาดข้อความที่ดึงมา
        cleaned_text = clean(text)
        # ลบอักขระที่ผิดพลาด
        final_text = clean_text(cleaned_text)
        # แสดงผลข้อความที่แก้ไขแล้ว
        print(final_text)
```
