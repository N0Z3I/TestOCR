```python
import fitz  # PyMuPDF
import pytesseract
from PIL import Image, ImageEnhance, ImageFilter
import io

# ตั้งค่าเส้นทางของ Tesseract (ถ้าจำเป็น)
# pytesseract.pytesseract.tesseract_cmd = r'/path/to/tesseract'

# โหลด PDF
pdf_path = 'Calendar_KPS_2567.pdf'
pdf_document = fitz.open(pdf_path)

# อ่านทุกหน้าของ PDF
for page_number in range(pdf_document.page_count):
    page = pdf_document.load_page(page_number)
    
    # แปลงหน้า PDF เป็นรูปภาพ (เพิ่ม DPI)
    pix = page.get_pixmap(dpi=100)
    img_data = pix.tobytes("png")
    img = Image.open(io.BytesIO(img_data))
    
    # แปลงภาพเป็นระดับสีเทา
    img = img.convert('L')
    
    # เพิ่มความคมชัด
    enhancer = ImageEnhance.Contrast(img)
    img = enhancer.enhance(2)

    # ใช้ Tesseract OCR ในการอ่านข้อความ
    custom_oem_psm_config = r'--oem 3 --psm 6'  # ใช้ OCR Engine Mode 3 (default) และ Page Segmentation Mode 6 (สำหรับบล็อคข้อความ)
    text = pytesseract.image_to_string(img, lang='tha+eng', config=custom_oem_psm_config)
    
    print(f"ข้อความจากหน้า {page_number + 1}:")
    print(text)

pdf_document.close()
```
